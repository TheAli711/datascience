# K-Means Clustering Code in Python

```python
import random
# Generating random observation values
import numpy as np
import matplotlib.pyplot as plt
import time


# Creates a numpy array of shape (1,10), filled with random integers between 0 (inclusive) and 50 (exclusive)
observations = np.random.randint(0,20,(1,20))

# Create n clusters
k = int(input("How many cluster do you want(2-3)? "))
clusters = [[] for i in range(k)]
centroids = [random.randint(np.amin(observations), np.amax(observations) + 1) for i in range(k)]
prevCentroids = [0 for i in range(k)]
while(not(np.array_equal(centroids, prevCentroids))):
    colors = ['ro', 'go', 'yo', 'bo']
    tempClusters = [[] for i in range(k)]
    for x in observations[0]:
        distances = [np.absolute(x-j) for j in centroids]
        tempClusters[distances.index(min(distances))].append(x)
    clusters = tempClusters
    plt.figure()
    for i in clusters:
        plt.xlim([-10,30])
        plt.ylim([-10,30])
        plt.plot(i, colors.pop())
    plt.show
    prevCentroids = centroids
    centroids = [np.mean(i) for i in clusters]
```
